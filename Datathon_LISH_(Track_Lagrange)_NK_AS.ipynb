# **Data Challenge Overview -- Pump it Up: Data Mining the Water Table**

A smart understanding of which waterpoints will fail can improve maintenance operations and ensure that clean, potable water is available everywhere.  In this competition, you will develop a predictive model that solve binary classification task focused on predicting the operational status of water pumps throughout Tanzania, based on some the provided information about their installation context. 

We will evaluate prediction quality according to a [Log-Loss Error Function](https://www.kaggle.com/dansbecker/what-is-log-loss). (Note that this is different than Percent Accuracy!) As you know, a good statistical model is paramount for helping organizations allocate maintenance resources effectively. ### Feature Engineering

```python
# Loading testing data to conduct feature engineering on train and test together

# Download file with features in the test set
testFeaturesURL = "https://drive.google.com/uc?id=1C64MAXqe4R4UjFBJcyK7hmKcNIVj3C8x"
gdown.download(testFeaturesURL, "test_features.csv", quiet=True)

# Read test set features into a pandas data frame
testFeatures = pd.read_csv("test_features.csv")


testFeatures["date_recorded"] = pd.to_datetime(testFeatures["date_recorded"])
testFeatures["year_recorded"] = pd.DatetimeIndex(testFeatures.date_recorded).year
testFeatures["month_recorded"] = pd.DatetimeIndex(testFeatures.date_recorded).month
testFeatures["day_recorded"] = pd.DatetimeIndex(testFeatures.date_recorded).day
testFeatures["years_since_created"] = testFeatures.year_recorded - testFeatures.construction_year
testFeatures = testFeatures.drop(columns = ['id','date_recorded','recorded_by'])
testFeatures
```

```python
# Feature engineering for training and testing data

trainx = train.drop(columns=['status_group'])
trainy = train.status_group

num_cols = trainx._get_numeric_data().columns
cat_cols = list(set(trainx.columns) - set(num_cols))

temp = pd.concat([trainx,testFeatures])

temp_cat = temp[cat_cols]
temp_num = temp[num_cols]


oe = OrdinalEncoder()
cols = temp_cat.columns
temp_cat = oe.fit_transform(temp_cat)
temp_cat = pd.DataFrame(temp_cat,columns=cols)

temp_num.reset_index(inplace=True, drop=True)

temp = pd.concat([temp_num,temp_cat],axis=1)

for col in cat_cols:
  try:
    temp[col] = temp[col].astype('str')
  except:
    continue
for col in num_cols:
  try:
    temp[col] = temp[col].astype('int')
  except:
    continue

tempx = temp.iloc[:len(trainx),:]
test = temp.iloc[len(trainx):,:]
```

```python
#Viewing the final training features
tempx
```

```python
#Checking train data column types
tempx.dtypes
```

### Model Building

```python
# Loading model (Best score on public)
cat = CatBoostClassifier(allow_writing_files=False,learning_rate = 0.07,depth=8)
```

```python
#One of the two models submitted
#cat = CatBoostClassifier(allow_writing_files=False)
```

### Model Evaluation

```python
#Conducting cross validation and predicting for test data
k=10
skf = StratifiedKFold(n_splits=k,
                      shuffle=True,
                      random_state=int(time.time()))

list_of_train_index = []
list_of_valid_index = []

for train_index, valid_index in skf.split(np.zeros(len(trainy)), # Placeholder (dummy input)
                                          trainy): # True input
    
    list_of_train_index.append(train_index)
    list_of_valid_index.append(valid_index)
loss = []
probabs_test = np.zeros([len(test),k])
for j in range(k):
            
    x_train = tempx.loc[list_of_train_index[j], :]
    x_valid = tempx.loc[list_of_valid_index[j], :]
    
    y_true_train = trainy[list_of_train_index[j]]
    y_true_valid = trainy[list_of_valid_index[j]]

    cat.fit(x_train,y_true_train,cat_features=cat_cols)
    probabs = cat.predict_proba(x_valid)[:,1]
    probabs_test[:,j] = cat.predict_proba(test)[:,1]
    loss.append(log_loss(y_true_valid,probabs))

loss
```

```python
# Creating an ensemble for prediction
preds = probabs_test.mean(axis=1)
```

```python
# Mean validation log-loss
np.mean(loss)
```
